```python
import itertools
import random

GRID_SIZE = 3
NUM_COUNT = 4
DIGITS = '0123456789'

# -----------------------------
# í›„ë³´ ìƒíƒœ ìƒì„±
# -----------------------------
def generate_possible_states():
    states = []

    positions = [(r, c) for r in range(GRID_SIZE) for c in range(GRID_SIZE)]

    for digits in itertools.permutations(DIGITS, NUM_COUNT):
        for pos_combo in itertools.permutations(positions, NUM_COUNT):
            state = tuple((digits[i], pos_combo[i][0], pos_combo[i][1])
                          for i in range(NUM_COUNT))
            states.append(state)

    return states


# -----------------------------
# íŒì • í•¨ìˆ˜ (S / B / I)
# -----------------------------
def evaluate_guess(secret, guess):
    S = B = I = 0

    secret_map = {d: (r, c) for d, r, c in secret}

    for d, gr, gc in guess:
        if d not in secret_map:
            continue

        sr, sc = secret_map[d]

        if gr == sr and gc == sc:
            S += 1
        elif gc == sc or gr == sr:
            B += 1
        else:
            I += 1

    return S, B, I


# -----------------------------
# í›„ë³´ í•„í„°ë§
# -----------------------------
def filter_possible_states(states, guess, S, B, I):
    remaining = []
    for state in states:
        s, b, i = evaluate_guess(state, guess)
        if (s, b, i) == (S, B, I):
            remaining.append(state)
    return remaining


# -----------------------------
# Helper ë£¨í”„
# -----------------------------
def play_2d_baseball_helper():
    possible_states = generate_possible_states()
    print(f"Total possible states: {len(possible_states)}")

    while True:
        if not possible_states:
            print("âŒ No possible states left. Check inputs.")
            break

        if len(possible_states) == 1:
            print("\nğŸ‰ Solution Found!")
            print_state(possible_states[0])
            break

        suggestion = random.choice(possible_states)
        print("\nğŸ‘‰ Suggested guess:")
        print_state(suggestion)

        raw = input("Enter your guess as: digit,row,col | digit,row,col | ... (or q): ")
        if raw.lower() == 'q':
            break

        try:
            guess = []
            parts = raw.split('|')
            for p in parts:
                d, r, c = p.strip().split(',')
                guess.append((d, int(r), int(c)))

            if len(guess) != NUM_COUNT:
                raise ValueError
        except:
            print("âŒ Invalid guess format.")
            continue

        try:
            S = int(input("Enter S count: "))
            B = int(input("Enter B count: "))
            I = int(input("Enter I count: "))
        except:
            print("âŒ Invalid result input.")
            continue

        if S + B + I > NUM_COUNT:
            print("âŒ Invalid result total.")
            continue

        possible_states = filter_possible_states(
            possible_states, guess, S, B, I
        )

        print(f"Remaining states: {len(possible_states)}")


# -----------------------------
# ì¶œë ¥ìš©
# -----------------------------
def print_state(state):
    board = [['.' for _ in range(GRID_SIZE)] for _ in range(GRID_SIZE)]
    for d, r, c in state:
        board[r][c] = d
    for row in board:
        print(" ".join(row))


# -----------------------------
# ì‹¤í–‰
# -----------------------------
print("ğŸ¯ 2D Number Baseball Helper")
play_2d_baseball_helper()
```
